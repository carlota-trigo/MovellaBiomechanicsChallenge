```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_folder = 'C:/Users/Carlota/Documents/MovellaBiomechanicsChallenge/data/'
path_to_pkl = data_folder + 'P12/PKL/P12_easy.pkl'
with open(path_to_pkl, 'rb') as f:
    data = pickle.load(f)
```

```python
joint_names = list(data['segmentData'].keys())
```

```python
jerk_list = []
for joint in joint_names: 
    acceleration = data['segmentData'][joint]['acceleration']
    magnitude = np.linalg.norm(acceleration, axis=1)
    jerk = np.diff(magnitude)
    jerk_list.append(jerk)
```

```python
# Create a figure with 23 subplots arranged in a grid with 5 rows and 5 columns
fig, axs = plt.subplots(6, 4, figsize=(15, 15))

# Flatten the 2D array of subplots into a 1D array for easier indexing
axs = axs.flatten()

# Generate some random data for each subplot
for i in range(23):
    # Generate random data for x and y axes
    axs[i].plot(jerk_list[i])
    
    # Set the title and plot the data in the current subplot
    axs[i].set_title(joint_names[i])    
    
# Hide any unused subplots
for i in range(23, len(axs)):
    axs[i].set_visible(False)

# Add a main title to the figure
fig.suptitle('Jerk in Each Segment')

# Adjust the spacing between subplots
fig.tight_layout()

# Show the figure
plt.show()
```

```python
rms_jerk_list = []
for i in range(23):
    # Compute the root-mean-square of jerk to get a measure of smoothness
    rms_jerk = np.sqrt(np.mean(np.square(jerk_list[i])))
    rms_jerk_list.append(rms_jerk)
    #print('RMS jerk ' + joint_names[i] +': ', rms_jerk)
plt.barh(joint_names, rms_jerk_list)
```
